#QuickSort

```python

def quicksort(L: list[int]) -> None:
    
    def splitsort(L: list[int], first:int, last:int) -> None:
        if first < last:
            slot = first
            for i in range(first, last):
                if L[i] < L[last]:
                    L[slot], L[i] = L[i], L[slot]
                    slot += 1
            L[slot], L[last] = L[last], L[slot]
            splitsort(L, first, slot-1)
            splitsort(L, slot+1, last)

    splitsort(L, 0, len(L)-1)
```

```python
lists = [[], [1], [1,2], [2,1], [1,2,3,4,5], [5,4,3,2,1], [4,2,3,1,5]]
for l in lists:
    quicksort(l)
    print(l)
```
