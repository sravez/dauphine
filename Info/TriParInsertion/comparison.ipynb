```python
import time
import matplotlib.pyplot as plt
import random
import statistics
```

# Comparaison de performance des fonctions de tri par insertion

## Fonction par descente de bulle

```python
def insertion_sort(l: list[int]):
    for k in range(1, len(l)):
        val = l[k]
        j = k
        while j > 0 and val < l[j-1]:
            l[j] = l[j-1]
            j -= 1
        # Inutile si j=k mais pas forcément plus coûteux qu'un test systématique
        l[j] = val

def insertion_sort_test(l: list[int]):
    for k in range(1, len(l)):
        val = l[k]
        j = k
        while j > 0 and val < l[j-1]:
            l[j] = l[j-1]
            j -= 1
        if(j != k):
            l[j] = val
```

### Tests élémentaires

```python
lists = [ [], [1], [1,2,3,4,5], [5,4,3,2,1], [5,2,4,1,3] ]
for L in lists:
    print(L, end=" -> ")
    insertion_sort(L)
    print(L)

lists = [ [], [1], [1,2,3,4,5], [5,4,3,2,1], [5,2,4,1,3] ]
for L in lists:
    print(L, end=" -> ")
    insertion_sort_test(L)
    print(L)

l = [ random.randint(0,500) for _ in range(100) ]
insertion_sort_test(l)
print(l)    
```

## Fonction par échange

```python
def insertion_sort_swap(l: list[int]):
    for k in range(1, len(l)):
        j = k
        while j > 0 and l[k] < l[j-1]:
            j -= 1
        l.insert(j, l.pop(k))
            
def insertion_sort_swap_test(l: list[int]):
    for k in range(1, len(l)):
        j = k
        while j > 0 and l[k] < l[j-1]:
            j -= 1
        if(j != k):
            l.insert(j, l.pop(k))
            
```

### Tests élémentaires

```python
lists = [ [], [1], [1,2,3,4,5], [5,4,3,2,1], [5,2,4,1,3] ]
for L in lists:
    print(L, end=" -> ")
    insertion_sort_swap(L)
    print(L)

lists = [ [], [1], [1,2,3,4,5], [5,4,3,2,1], [5,2,4,1,3] ]
for L in lists:
    print(L, end=" -> ")
    insertion_sort_swap_test(L)
    print(L)

l = [ random.randint(0,100) for _ in range(100) ]
print(l)
insertion_sort_swap_test(l)
print(l)
```

## Temps de calculs

```python
list_sizes = range(100,1001)
sample_size = 10
max_val = int(1e6)
```

```python
# Temps moyens pour chaque taille de liste
times = []
times_test = []
times_swap = []
times_swap_test = []

for n in list_sizes:
    sample = []
    sample_test = []
    sample_swap = []
    sample_swap_test = []

    for i in range(sample_size):
        l = [random.randint(0, max_val) for _ in range(n)]

        l_copy = l.copy()
        t = time.monotonic()
        insertion_sort(l_copy)   
        sample.append(time.monotonic() - t)
        
        l_copy = l.copy()
        t = time.monotonic()
        insertion_sort_test(l_copy)
        sample_test.append(time.monotonic() - t)
        
        l_copy = l.copy()
        t = time.monotonic()
        insertion_sort_swap(l_copy)
        sample_swap.append(time.monotonic() - t)
        
        t = time.monotonic()
        insertion_sort_swap_test(l)
        sample_swap_test.append(time.monotonic() - t)

    times.append(statistics.mean(sample))
    times_test.append(statistics.mean(sample_test))
    times_swap.append(statistics.mean(sample_swap))
    times_swap_test.append(statistics.mean(sample_swap_test))
```

### Comparaison std vs std_test

```python
plt.plot(list_sizes, times, label="Std")
plt.plot(list_sizes, times_test, label="Std_Test")
plt.legend()
```

### Comparaison swap vs swap_test

```python
plt.plot(list_sizes, times_swap, label="swap")
plt.plot(list_sizes, times_swap_test, label="swap Test")
plt.legend()
```

### Comparaison std_test vs swap_test

```python
plt.plot(list_sizes, times_test, label="Std Test")
plt.plot(list_sizes, times_swap_test, label="swap Test")
plt.legend()
```
